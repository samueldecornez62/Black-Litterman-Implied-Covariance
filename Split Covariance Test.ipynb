```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Create a random covariance matrix
np.random.seed(42)  # For reproducibility
data = np.random.randn(200, 200)
cov_matrix = np.cov(data, rowvar=False)

# Step 2: Split the covariance matrix into quarters
top_left = cov_matrix[:100, :100]
top_right = cov_matrix[:100, 100:]
bot_left = cov_matrix[100:, :100]
bot_right = cov_matrix[100:, 100:]

# Step 3: Plot the quarters in a single output
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Top Left
sns.heatmap(top_left, ax=axs[0, 0], cmap='viridis', annot=False, linewidths=.5)
axs[0, 0].set_title('Top Left', fontsize=14)

# Top Right
sns.heatmap(top_right, ax=axs[0, 1], cmap='viridis', annot=False, linewidths=.5)
axs[0, 1].set_title('Top Right', fontsize=14)

# Bottom Left
sns.heatmap(bot_left, ax=axs[1, 0], cmap='viridis', annot=False, linewidths=.5)
axs[1, 0].set_title('Bottom Left', fontsize=14)

# Bottom Right
sns.heatmap(bot_right, ax=axs[1, 1], cmap='viridis', annot=False, linewidths=.5)
axs[1, 1].set_title('Bottom Right', fontsize=14)

plt.tight_layout()
plt.suptitle('Heatmaps of Covariance Matrix Quarters', fontsize=16)
plt.subplots_adjust(top=0.9)  # Adjust title position
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Create a random covariance matrix
np.random.seed(42)  # For reproducibility
data = np.random.randn(200, 200)
cov_matrix = np.cov(data, rowvar=False)

# Step 2: Split the covariance matrix into quarters
top_left = cov_matrix[:100, :100]
top_right = cov_matrix[:100, 100:]
bot_left = cov_matrix[100:, :100]
bot_right = cov_matrix[100:, 100:]

# Step 3: Determine the color scale limits based on the full covariance matrix
vmin = cov_matrix.min()
vmax = cov_matrix.max()

# Step 4: Plot the original covariance matrix and the quarters with consistent colors
fig, axs = plt.subplots(3, 2, figsize=(12, 15))

# Full Original Covariance Matrix
sns.heatmap(cov_matrix, ax=axs[0, 0], cmap='viridis', annot=False, linewidths=.5, vmin=vmin, vmax=vmax)
axs[0, 0].set_title('Full Covariance Matrix', fontsize=16)
axs[0, 0].set_xticks([])  # Remove x ticks
axs[0, 0].set_yticks([])  # Remove y ticks

# Top Left Quarter
sns.heatmap(top_left, ax=axs[1, 0], cmap='viridis', annot=False, linewidths=.5, vmin=vmin, vmax=vmax)
axs[1, 0].set_title('Top Left', fontsize=14)

# Top Right Quarter
sns.heatmap(top_right, ax=axs[1, 1], cmap='viridis', annot=False, linewidths=.5, vmin=vmin, vmax=vmax)
axs[1, 1].set_title('Top Right', fontsize=14)

# Bottom Left Quarter
sns.heatmap(bot_left, ax=axs[2, 0], cmap='viridis', annot=False, linewidths=.5, vmin=vmin, vmax=vmax)
axs[2, 0].set_title('Bottom Left', fontsize=14)

# Bottom Right Quarter
sns.heatmap(bot_right, ax=axs[2, 1], cmap='viridis', annot=False, linewidths=.5, vmin=vmin, vmax=vmax)
axs[2, 1].set_title('Bottom Right', fontsize=14)

plt.tight_layout()
plt.suptitle('Heatmap of Covariance Matrix and its Quarters', fontsize=18)
plt.subplots_adjust(top=0.93)  # Adjust title position
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Create a random covariance matrix
np.random.seed(42)  # For reproducibility
data = np.random.randn(200, 200)
cov_matrix = np.cov(data, rowvar=False)

# Step 2: Split the covariance matrix into quarters
top_left = cov_matrix[:100, :100]
top_right = cov_matrix[:100, 100:]
bot_left = cov_matrix[100:, :100]
bot_right = cov_matrix[100:, 100:]

# Step 3: Determine the color scale limits based on the full covariance matrix
vmin = cov_matrix.min()
vmax = cov_matrix.max()

# Step 4: Plot the quarters first, then the full covariance matrix
fig, axs = plt.subplots(3, 2, figsize=(12, 15))

# Top Left Quarter
sns.heatmap(top_left, ax=axs[0, 0], cmap='viridis', annot=False, linewidths=.5, vmin=vmin, vmax=vmax)
axs[0, 0].set_title('Top Left', fontsize=14)

# Top Right Quarter
sns.heatmap(top_right, ax=axs[0, 1], cmap='viridis', annot=False, linewidths=.5, vmin=vmin, vmax=vmax)
axs[0, 1].set_title('Top Right', fontsize=14)

# Bottom Left Quarter
sns.heatmap(bot_left, ax=axs[1, 0], cmap='viridis', annot=False, linewidths=.5, vmin=vmin, vmax=vmax)
axs[1, 0].set_title('Bottom Left', fontsize=14)

# Bottom Right Quarter
sns.heatmap(bot_right, ax=axs[1, 1], cmap='viridis', annot=False, linewidths=.5, vmin=vmin, vmax=vmax)
axs[1, 1].set_title('Bottom Right', fontsize=14)

# Full Original Covariance Matrix
sns.heatmap(cov_matrix, ax=axs[2, 0], cmap='viridis', annot=False, linewidths=.5, vmin=vmin, vmax=vmax)
axs[2, 0].set_title('Full Covariance Matrix', fontsize=16)
axs[2, 0].set_xticks([])  # Remove x ticks
axs[2, 0].set_yticks([])  # Remove y ticks
axs[2, 1].axis('off')  # Hide the unused subplot

plt.tight_layout()
plt.suptitle('Heatmap of Covariance Matrix and its Quarters', fontsize=18)
plt.subplots_adjust(top=0.93)  # Adjust title position
plt.show()
```
